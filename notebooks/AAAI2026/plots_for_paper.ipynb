### Code for plots to put in AAAI paper

```python
import matplotlib.pyplot as plt
import numpy as np


datasets = ['ETTh1', 'ETTh2', 'ETTm1', 'ETTm2', 'Exchange Rate', 'Traffic']
models = ['ZS', 'FFT', 'Our approach (using ZS)', 'Our approach (using FFT)']

mse_values = np.array([
    [0.0524, 0.0512, 0.0439, 0.0442],   
    [0.1306, 0.1304, 0.1002, 0.0992],   
    [0.0275, 0.0264, 0.0221, 0.0215],   
    [0.0765, 0.0646, 0.0644, 0.0658],  
    [0.0790, 0.0883, 0.0492, 0.0484],   
    [0.1766, 0.1173, 0.1407, 0.0933],    
])

# Create bar positions
x = np.arange(len(datasets))  
bar_width = 0.18

# Colors for models
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#9467bd']

fig, ax = plt.subplots(figsize=(9, 5))

# Plot bars for each model
for i, model in enumerate(models):
    ax.bar(x + i * bar_width, mse_values[:, i], width=bar_width, label=model, color=colors[i])

# Axis labels and ticks
ax.set_xlabel('Dataset Name', fontsize=12)
ax.set_ylabel('MSE', fontsize=12)
# ax.set_title('Performance Comparison of Selective Forecasing using shapelets against baselines ', fontsize=13, pad=15)
ax.set_xticks(x + bar_width * (len(models) - 1) / 2)
ax.set_xticklabels(datasets, rotation=0, ha='center')

# Legend and grid
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Optional: adjust layout
plt.tight_layout()
plt.show()
```
